```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("~/.config/matplotlib/stylelib/rotskoff.mpl")
from mpl_toolkits.mplot3d import Axes3D

from matplotlib.colors import LinearSegmentedColormap
zissou = LinearSegmentedColormap.from_list(
    "zissou", ['#01abe9', '#f1f8f1', '#f54b1a'])

#plt.register_cmap(name="zissou", cmap=zissou)

def plot_gaussian_surface_with_black_arrows():
    # Define the range and meshgrid for x and y
    x = np.linspace(-3, 3, 1000)
    y = np.linspace(-3, 3, 1000)
    x, y = np.meshgrid(x, y)

    # Define the Gaussian function
    sigma = 1
    z = (1 / (2 * np.pi * sigma ** 2)) * np.exp(-(x**2 + y**2) / (2 * sigma**2))

    # Plot the surface
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, rstride=5, cstride=5, cmap=zissou, linewidth=0.)

    # Hide axes
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.axis('off')

    # # Adding black arrows as a schematic on top of the figure
    # # Arrow 1
    # ax.quiver(0, 0, z[50, 50], 0, 0, 1.5, color='black', arrow_length_ratio=.01)

    # # Arrow 2
    # ax.quiver(3, 0, z[50, 70], 1.5, 0, 0, color='black', arrow_length_ratio=0.1)

    # # Arrow 3
    # ax.quiver(-3, 0, z[50, 30], -1.5, 0, 0, color='black', arrow_length_ratio=0.1)

    fig.savefig("gaussian_base.pdf")
    plt.show()

# Call the function to plot the Gaussian distribution with black arrows
plot_gaussian_surface_with_black_arrows()
```
